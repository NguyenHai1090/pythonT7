```python
from tkinter import *
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk

class NhanSu:
    def __init__(self, HoTen, QueQuan, PhongBan, Luong, NgayVao):
        self.HoTen = HoTen
        self.QueQuan = QueQuan
        self.PhongBan = PhongBan
        self.Luong = Luong
        self.NgayVao = NgayVao

# tạo cửa sổ ứng dụng
window = Tk()

# setup thông tin cho window
window.title("Hệ Thống Quản lý Nhân sự Công ty TNHH Bình Minh")
window.geometry("500x150")

# tạo list để lưu trữ toàn bộ các object
list_NhanSu = []

# Setup sổ treeview và tiêu đề
root = tk.Tk()
root.title('Dữ liệu Nhân sự Công ty TNHH Bình Minh')
root.geometry("1000x200")

# Define cột treeview
columns = ("HoTen", "QueQuan", "PhongBan", "Luong", "NgayVao")

# Tạo widge treeview và tiêu đề cột
tree = ttk.Treeview(root, columns=columns, show='headings')
tree.heading('HoTen', text='Họ Và Tên')
tree.heading('QueQuan', text='Quê Quán')
tree.heading('PhongBan', text='Mã Phòng Ban')
tree.heading('Luong', text='Lương hiện tại')
tree.heading('NgayVao', text='Ngày vào công ty')
# Đặt widget treeview trên cửa số gốc
tree.grid(row=0, column=0, sticky='nsew')

# thêm thanh cuộn dọc ở cửa số treeview
scrollbar = ttk.Scrollbar(root, orient=tk.VERTICAL, command=tree.yview)
tree.configure(yscroll=scrollbar.set)
scrollbar.grid(row=0, column=1, sticky='ns')

# mô tả các function cho ứng dụng
# add/search/update/delete/reset

def Them_info():
    HoTen_value = HoTen.get()
    QueQuan_value = QueQuan.get()
    PhongBan_value = PhongBan.get()
    Luong_value = int(Luong.get())
    NgayVao_value = NgayVao.get()
    new_NhanSu = NhanSu(HoTen_value, QueQuan_value, PhongBan_value, Luong_value, NgayVao_value)
    list_NhanSu.append(new_NhanSu)
    # xóa giá trị ở các entry
    HoTen.delete(0, END)
    QueQuan.delete(0, END)
    PhongBan.delete(0, END)
    Luong.delete(0, END)
    NgayVao.delete(0, END)
    # Xóa toàn bộ dữ liệu trong Treeview
    for item in tree.get_children():
        tree.delete(item)
    # Thêm dữ liệu mới vào Treeview
    for nhansu in list_NhanSu:
        tree.insert('', tk.END, values=(nhansu.HoTen, nhansu.QueQuan, nhansu.PhongBan, nhansu.Luong, nhansu.NgayVao))
    messagebox.showinfo("Thêm info", f"{HoTen_value} is added successfully!")

def TimKiem_info():
    TimKiem_name_value = TimKiem_name.get()
    is_exist = False
    for nhansu in list_NhanSu:
        if nhansu.HoTen == TimKiem_name_value:
            messagebox.showinfo("Tìm Kiếm nhân sự", f"Họ và Tên: {nhansu.HoTen}, Quê Quán: {nhansu.QueQuan}, Phòng Ban: {nhansu.PhongBan}, Lương :{nhansu.Luong}, Ngày vào công ty: {nhansu.NgayVao}")
            is_exist = True
    if not is_exist:
        messagebox.showinfo("Tìm Kiếm nhân sự", "Không tìm thấy nhân sự phù hợp!")

def CapNhat_info():
    messagebox.showinfo("Cập nhật nhân sự", "Hệ thống chỉ cập nhật tiền Lương")
    TimKiem_name_value = TimKiem_name.get()
    Luong_value = Luong.get()
    is_exist = False
    for i in range(len(list_NhanSu)):
        if list_NhanSu[i].HoTen == TimKiem_name_value:
            list_NhanSu[i].Luong = Luong_value
            is_exist = True
            messagebox.showinfo("Cập nhật nhân sự", f"Cập nhật tiền lương cho nhân sự: {TimKiem_name_value} thành công!")
            break
    if not is_exist:
        messagebox.showinfo("Cập nhật nhân sự", "Không tìm thấy nhân sự phù hợp!")
    # Xóa toàn bộ dữ liệu trong Treeview
    for item in tree.get_children():
        tree.delete(item)
    # Thêm dữ liệu mới vào Treeview
    for nhansu in list_NhanSu:
        tree.insert('', tk.END, values=(nhansu.HoTen, nhansu.QueQuan, nhansu.PhongBan, nhansu.Luong, nhansu.NgayVao))
        
def Xoa_info():
    TimKiem_name_value = TimKiem_name.get()
    is_exist = False
    for i in range(len(list_NhanSu)):
        if list_NhanSu[i].HoTen == TimKiem_name_value:
            del list_NhanSu[i]
            is_exist = True
            messagebox.showinfo("Xóa nhân sự", f"Xóa nhân sự: {TimKiem_name_value} thành công!")
            break
    if not is_exist:
        messagebox.showinfo("Xóa nhân sự", "Không tìm thấy nhân sự cần xóa!")
    # Xóa toàn bộ dữ liệu trong Treeview
    for item in tree.get_children():
        tree.delete(item)
    # Thêm dữ liệu mới vào Treeview
    for nhansu in list_NhanSu:
        tree.insert('', tk.END, values=(nhansu.HoTen, nhansu.QueQuan, nhansu.PhongBan, nhansu.Luong, nhansu.NgayVao))
        
def CatDatLai_info():
    HoTen.delete(0, END)
    QueQuan.delete(0, END)
    PhongBan.delete(0, END)
    Luong.delete(0, END)
    NgayVao.delete(0, END)
    # Xóa toàn bộ dữ liệu trong Treeview
    for item in tree.get_children():
        tree.delete(item)
    # Thêm dữ liệu mới vào Treeview
    for nhansu in list_NhanSu:
        tree.insert('', tk.END, values=(nhansu.HoTen, nhansu.QueQuan, nhansu.PhongBan, nhansu.Luong, nhansu.NgayVao))
# setup các widget

# tạo label
HoTen_lbl = Label(window, text="Họ và Tên:")
QueQuan_lbl = Label(window, text="Quê Quán:")
PhongBan_lbl = Label(window, text="Mã Phòng Ban:")
Luong_lbl = Label(window, text="Lương hiện tại:")
NgayVao_lbl = Label(window, text="Ngày vào công ty:")

# tạo entry
HoTen = Entry(window)
QueQuan = Entry(window)
PhongBan = Entry(window)
Luong = Entry(window)
NgayVao = Entry(window)
TimKiem_name = Entry(window)

# tạo button
Them_btn = Button(window, text="Thêm", command=Them_info)
TimKiem_btn = Button(window, text="Tìm Kiếm", command=TimKiem_info)
CapNhat_btn = Button(window, text="Cập Nhật", command=CapNhat_info)
Xoa_btn = Button(window, text="Xóa", command=Xoa_info)
CatDatLai_btn = Button(window, text="Cài Đặt Lại", command=CatDatLai_info)

# dán các widget lên window
HoTen_lbl.grid(row=0, column=0)
HoTen.grid(row=0, column=1)
QueQuan_lbl.grid(row=1, column=0)
QueQuan.grid(row=1, column=1)
PhongBan_lbl.grid(row=2, column=0)
PhongBan.grid(row=2, column=1)
Luong_lbl.grid(row=0, column=2)
Luong.grid(row=0, column=3)
NgayVao_lbl.grid(row=1, column=2)
NgayVao.grid(row=1, column=3)
Them_btn.grid(row=3, column=0)
Xoa_btn.grid(row=2, column=2)
TimKiem_btn.grid(row=3, column=2)
CapNhat_btn.grid(row=2, column=3)
TimKiem_name.grid(row=3, column=1)
CatDatLai_btn.grid(row=3, column=3)

# hiển thị cửa sổ root
root.mainloop()

# chạy ứng dụng
window.mainloop()
```
